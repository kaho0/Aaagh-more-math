```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({'h':[150,160,170,180,190],
                 'w':[50,60,70,80,90]})
df
```

```python
mean=df.mean()
s=df.std()
print(mean)
print(s)
z=(df-mean)/s
z.round(2)
```

```python
z.plot(kind='bar')
plt.title('Z-score')
plt.xlabel('Index')
plt.ylabel('z values')
plt.show()
```
