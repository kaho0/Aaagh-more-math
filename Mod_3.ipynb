```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({'h':[150,160,170,180,190],
                 'w':[50,60,70,80,90]})
```

```python
mean=df.mean()
s=df.std()
print(mean)
print(s)
z=(df-mean)/s
z.round(2)
```

```python
z.plot(kind='bar')
plt.title('Z-score')
plt.xlabel('Index')
plt.ylabel('z values')
plt.show()
```

Minâ€“Max Scaling (Rescaling to [0,1])

```python
mn=df.min()
mx=df.max()
rg=mx-mn
print(mn,mx,rg)
```

```python
ss=df-mn
ss.head()
```

```python
mn=ss/rg
mn.round(2)
```

```python

#Plot
mn.plot(kind='bar')
plt.title("Min-Max Scaling [0-1]")
plt.xlabel("Index"); plt.ylabel('Min-max value')
plt.show()
```
