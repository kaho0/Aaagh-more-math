```python
target_col = "HeartDisease"

numeric_cols = ["Age", "RestingBP", "Cholesterol", "FastingBS",
                "MaxHR", "Oldpeak"]

categorical_cols = ["Sex", "ChestPainType", "RestingECG",
                    "ExerciseAngina", "ST_Slope"]

print("Target:", target_col)
print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)
```

```python
df['Age'].hist(bins=20,figsize=(8,4))
plt.title("Histogram of numerical Feature age ",fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df['RestingBP'].hist(bins=20,figsize=(8,4))
plt.title("Histogram of numerical Resting BP ",fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df['Cholesterol'].hist(bins=20,figsize=(8,4))
plt.title("Histogram of numerical Feature cholesterol ",fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df[numeric_cols].hist(bins=20,figsize=(8,4))
plt.title("Histogram of numerical Featured ",fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
df[numeric_cols].boxplot()
plt.title("Boxplot of numerical Features ",fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
df.boxplot(column=numeric_cols[0])
plt.title("Boxplot of numerical Features ",fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=df[target_col])
plt.title("Heart Disease class distribution ")
plt.xlabel("Heart Disease(0=No,1=Yes)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
for c in categorical_cols:
  plt.figure(figsize=(15,10))
  df[c].value_counts().plot(kind='bar')
  plt.title(f"Distribution of {c} ")
  plt.ylabel("Count")
  plt.tight_layout()
  plt.show()
```

```python
plt.figure(figsize=(15,8))
for i,col in enumerate(numeric_cols,1):
  plt.subplot(2,3,i)
  sns.boxplot(x=df[target_col],y=df[col])
  plt.title(col)
plt.tight_layout()
plt.show()
```
